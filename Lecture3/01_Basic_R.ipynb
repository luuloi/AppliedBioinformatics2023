**1.6 Summary Functions and NA**

```r
y <- c(NA, 7, 5, 6, 6, 8, 5)
```

```r
min(y)
```

```r
min(y, na.rm = TRUE)
```

```r
max(y, na.rm = TRUE)
```

```r
mean(y, na.rm = TRUE)
```

```r
median(y, na.rm = TRUE)
```
